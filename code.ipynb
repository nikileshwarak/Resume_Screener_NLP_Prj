## Cleaning Data

1. URL'S
2. Hashtags
3. mentions
4. special letters
5. punctuation

```python
import re

def cleanResume(txt):
    cleanText = re.sub(r'http\S+\s', ' ', txt)
    cleanText = re.sub(r'RT|cc', ' ', cleanText) #Removes "RT" (retweet) and "cc" (carbon copy mention)
    cleanText = re.sub(r'#\S+\s', ' ', cleanText)
    cleanText = re.sub(r'@\S+', '  ', cleanText)  
    cleanText = re.sub(r'[%s]' % re.escape(r"""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', cleanText)
    cleanText = re.sub(r'[^\x00-\x7f]', ' ', cleanText) #Match any character except standard ASCII characters (i.e., match non-English, emoji, symbols, etc.)
    cleanText = re.sub(r'\s+', ' ', cleanText)
    return cleanText
```

```python
cleanResume("This is my website https://helloworld.com and my gmail is @gmail.com ")
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

le.fit(df['Category'])

df['Category'] = le.transform(df['Category'])
y = df['Category']
```

```python
df
```

```python
df.Category.unique()
```

## Vectorization

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english') #vectorizer automatically remove common English stop words

x = tfidf.fit_transform(df['Resume']) # fit_transforms text into a sparse numerical matrix of TF-IDF features.
```

## Training

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

## Prediction

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

# 1. Train KNeighborsClassifier
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
print(accuracy_score(y_test, y_pred))
```

## Prediction System

```python
# Function to predict the category of a resume
def pred(input_resume):
    # Preprocess the input text (e.g., cleaning, etc.)
    cleaned_text = cleanResume(input_resume) 

    # Vectorize the cleaned text using the same TF-IDF vectorizer used during training
    vectorized_text = tfidf.transform([cleaned_text])
    
     # Prediction
    predicted_category = clf.predict(vectorized_text)

    # get name of predicted category
    predicted_category_name = le.inverse_transform(predicted_category) #inverse_transform maps those predicted numbers back to readable category names.

    return predicted_category_name[0]  # only want the first and only result, not a list
```
